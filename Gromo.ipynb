```python
#Joinning Order_Data and Customer_Data for further analysis.
DF = Order_data.merge(Customer_data,left_on='Cust ID',right_on='Cust Id',how ='left',indicator=True)
DF.head()
```

```python
unique_customers_placed_order = DF.groupby(['Order_Month', 'Cust ID'])['Order No'].count().reset_index()
unique_customers_placed_order = unique_customers_placed_order.groupby('Order_Month')['Cust ID'].nunique().reset_index()
unique_customers_placed_order = unique_customers_placed_order.rename(columns={'Cust ID': 'Unique Customers Placed Order'})

#Finding unique customers whose orders got converted each month
won_orders = DF[DF['Order Status'] == 'Won']
unique_customers_converted = won_orders.groupby(['Order_Month', 'Cust ID'])['Order No'].count().reset_index()
unique_customers_converted = unique_customers_converted.groupby('Order_Month')['Cust ID'].nunique().reset_index()
unique_customers_converted = unique_customers_converted.rename(columns={'Cust ID': 'Unique Customers Converted'})

#Merging the two datasets on 'Order_Month'
result = pd.merge(unique_customers_placed_order, unique_customers_converted, on='Order_Month', how='outer').fillna(0)


print("Unique Customers Placed Order and Converted Month-on-Month:")
print(result)


```

```python
# Bifurcate by Cust Profession
result_bifurcated = DF.groupby(['Order_Month', 'Cust Profession'])['Cust ID'].nunique().reset_index()
result_bifurcated = result_bifurcated.pivot(index='Order_Month', columns='Cust Profession', values='Cust ID').fillna(0)


print("\nBifurcated by Cust Profession:")
print(result_bifurcated)

#Transferring data bifurcated data to CSV
result_bifurcated.to_csv('/content/drive/MyDrive/CSV/plan.csv', index=False)

```

```python
# Converting 'Order Date' and 'Customer Created Date' to datetime format

DF['Order Date'] = pd.to_datetime(DF['Order Date'])
DF['Customer Created Date'] = pd.to_datetime(DF['Customer Created Date'])

# Extract the month and year from 'Order Date' for cohort grouping
DF['Order_Month'] = DF['Order Date'].dt.to_period('M')

# Extract the month and year from 'Customer Created Date' for cohort grouping
DF['Cohort_Month'] = DF['Customer Created Date'].dt.to_period('M')

# Create a cohort index based on the month a customer was onboarded
DF['Cohort_Index'] = (DF['Order_Month'].dt.year - DF['Cohort_Month'].dt.year) * 12 + (DF['Order_Month'].dt.month - DF['Cohort_Month'].dt.month)

# Filter the data to only include orders from the month of onboarding and onwards
filtered_data = DF[DF['Cohort_Index'] >= 0]

# Create a pivot table to store the revenue for each cohort and month
cohort_pivot = pd.pivot_table(
    filtered_data,
    values='Order Amount',
    index='Cohort_Month',
    columns='Cohort_Index',
    aggfunc={'Order Amount': 'sum'}
).fillna(0)

# Display the cohort analysis as a table
print(cohort_pivot)

```

```python
# Create a heatmap using Seaborn
plt.figure(figsize=(16, 8))
sn.heatmap(cohort_pivot, cmap="RdYlGn", annot=True, fmt=".1f", linewidths=.5)
plt.title('Cohort Analysis Heatmap')
plt.show()
```
